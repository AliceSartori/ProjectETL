```python
#Import dependences

import pandas as pd
from sqlalchemy import create_engine
```

# Store CVS into a DataFrame

```python
#store the file into a variable 
gross_df = pd.read_csv("alice/gross_rent.csv", encoding='utf-8')
gross_df
```

```python
#call .info() to see how pandas interpret the data and see if we have NaNs values 
gross_df.info()
```

```python
#call convert type function 
new_gross_df=gross_df.convert_dtypes()
new_gross_df.info()
```

# Create a dataframe subset 

```python
new_gross_df=new_gross_df[['id','State_Name','State_ab','City','Zip_Code', 'Lat','Lon','Mean','Median','Stdev']]
new_gross_df
```

```python
renamed_df=new_gross_df.rename(columns={"State_Name": "state",
                            "Zip_Code":"zip_code",
                            "Lat": "mean_household_income",
                            "Lon": "std_household_income",
                            "Mean": "mean_gross_rent",
                            "Median": "median_gross_rent",
                            "Stdev": "std_gross_rent",
                            "State_ab":"state_ab",          
                            "City": "city"})
renamed_df
```

```python
#changing the datatype of the zip code as integer
renamed_df['zip_code'].astype(int)
```

```python
#Sort the data by ID and STATE column 
sorted_df=renamed_df.sort_values(['id', 'state'])
sorted_df
```

```python
#Check if there are any duplicate rows (There are no duplicate rows)
sorted_df.duplicated(subset=None, keep='first').sum()
```

# Connect to local database on Postgresq

```python
#Created the connection 
engine = create_engine('postgresql+psycopg2://alicesartori@localhost/Restaurant')
conn = engine.connect()
```

# Use pandas to load csv converted DataFrame into database 

```python
zip_code_df.to_sql('zip_code', con=engine, if_exists='append', index=False)
```

```python
#gross_rent
gross_rent.to_sql('gross_rent', con=engine, if_exists='append', index=False)
```

```python
#city df
city.to_sql('city', con=engine, if_exists='append', index=False)
```

```python
#state_df
state.to_sql('state', con=engine, if_exists='append', index=False)
```

```python
#household income
household_table.to_sql('household_table', con=engine, if_exists='append', index=False)
```
