```python
#Import dependeces 
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV into DataFrame

```python
#store the file into a variable 
gross_df = pd.read_csv("gross_rent.csv", encoding='utf-8')
gross_df
```

```python
#call .info() to see how pandas interpret the data and see if we have NaNs values 
gross_df.info()
```

```python
#call convert type function 
new_gross_df=gross_df.convert_dtypes()
new_gross_df.info()
```

# Create a dataframe subset 

```python
new_gross_df=new_gross_df[['id','State_Name','State_ab','City','Zip_Code', 'Lat','Lon','Mean','Median','Stdev']]
new_gross_df
```

# Rename the columns

```python
renamed_df=new_gross_df.rename(columns={"State_Name": "state",
                            "Zip_Code":"zip_code",
                            "Lat": "mean_household_income",
                            "Lon": "std_household_income",
                            "Mean": "mean_gross_rent",
                            "Median": "median_gross_rent",
                            "Stdev": "std_gross_rent",
                            "State_ab":"state_ab",          
                            "City": "city"})
```

```python
#changing
renamed_df['zip_code'].astype(int)
```

```python
#Sort the data by the state column (added the 'id' in a second moment, after uploading the data. That's why the order
#is not reflected below when the connection to postgresql gets created)
sorted_df=renamed_df.sort_values(['id', 'state'])
sorted_df
```

```python
#Check if there are any duplicate rows (There are no duplicate rows)
sorted_df.duplicated(subset=None, keep='first').sum()
```

```python
sorted_df.duplicated().any()
```

# Connect to local database on Postgresql

```python
#Created the connection 
engine = create_engine('postgresql+psycopg2://alicesartori@localhost/Rent_DB')
conn = engine.connect()
```

```python
#The first time you run this line of code there will be no data returned as you need:
# - to create a database on PGadmind
# - to create a schema to be able to enter data 
# - run the query below
pd.read_sql("SELECT * FROM gross_rent", conn)
```

# Check for tables

```python
engine.table_names()
```

# Use pandas to load csv converted DataFrame into database

```python
# Load clean dataframe into database. 
# The second time you run this code you receive an error stating the column were already entered.
sorted_df.to_sql('gross_rent', con=engine, if_exists='append', index=False)
```
